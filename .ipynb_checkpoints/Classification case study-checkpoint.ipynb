### Check Multicollinearity for Categorical features

* **A chi-squared test (also chi-square or χ2 test) is a statistical hypothesis test that is valid to perform when the test statistic is chi-squared distributed under the null hypothesis, specifically Pearson's chi-squared test**

* **A chi-square statistic is one way to show a relationship between two categorical variables.**

* **Here we test correlation of Categorical columns with Target column i.e ProdTaken**

```python
# We can not perform .corr() on the categorical features (like nyumerical features) to find the relationship between two categorical features, we use chi square test. 

# helps in finding the relationship or corelation of categorical features with the target feature
```

```python
from scipy.stats import chi2_contingency

chi2_test = []

for i in categorical_features:
    if chi2_contingency(pd.crosstab(df['ProdTaken'], df[i]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to reject Null Hypothesis')

result = pd.DataFrame(data=[categorical_features, chi2_test]).T

result.columns = ['Column', 'Hypothesis Result']

result
```

**Report** 
* Here Gender Column Fails to Rejects Null Hypothesis which means it doesn't correlate with target column.

##### pd.crosstab(df['ProdTaken'], df[i]):
This part of the expression creates a contingency table (also known as a cross-tabulation or frequency table) between the target variable 'ProdTaken' and the current categorical feature df[i]. The contingency table represents the frequency counts of different combinations of categories between the two variables.

##### chi2_contingency(...):
The chi2_contingency function from the scipy.stats module performs a chi-square test of independence on the contingency table created in the previous step. This test evaluates whether there is a statistically significant association between the two categorical variables. It returns a tuple containing several values, including the chi-square test statistic and the p-value.

#####  chi2_contingency(...)[1]:
This part extracts the second element of the tuple returned by chi2_contingency, which corresponds to the p-value of the chi-square test. The p-value represents the probability of observing the data (or more extreme results) under the assumption that the null hypothesis is true.

##### < 0.05:
This comparison checks if the p-value obtained from the chi-square test is less than 0.05, which is a common significance threshold used in hypothesis testing. If the p-value is less than 0.05, it suggests that there is strong evidence against the null hypothesis, indicating a statistically significant association between the two variables. Conversely, if the p-value is greater than or equal to 0.05, it suggests that there is insufficient evidence to reject the null hypothesis, indicating no significant association between the variables.

# EXTRA KNOWLEDGE 

enumerate() is a built-in Python function used to iterate over a sequence (such as a list, tuple, or string) while also keeping track of the index of each item in the sequence. It returns an enumerate object, which contains pairs of index and item values.

You can use enumerate() when you need to access both the index and the value of each element in a sequence within a loop. This is particularly useful in scenarios where you want to perform some operation based on both the index and the value, such as updating elements in the sequence or performing calculations based on their position.

```python
my_list = ['apple', 'banana', 'orange']

for index, value in enumerate(my_list):
    print(f"Index: {index}, Value: {value}")
```

In this example, enumerate(my_list) returns an enumerate object that yields tuples containing the index and value of each element in my_list. The for loop iterates over these tuples, unpacking them into index and value variables, which can then be used within the loop body.

You can use enumerate() in situations where you need to keep track of the index of elements in addition to their values, without having to manually manage the index variable yourself. It provides a cleaner and more concise way to achieve this functionality.

## Bias-variance tradeoff in Machine Learning is:

##### "A high-bias model will lead to underfitting, while a high-variance model tends to overfit the training data."

#### Explanation:

##### Bias:
Bias refers to the error introduced by approximating a real-world problem with a simplified model. A high-bias model makes strong assumptions about the underlying data and may not capture the true relationship between features and target variable. This can lead to underfitting, where the model performs poorly on both the training and unseen data.


##### Variance:
Variance refers to the variability of model predictions for different training datasets. A high-variance model is sensitive to small fluctuations in the training data and may capture noise rather than the underlying patterns. This can lead to overfitting, where the model performs well on the training data but poorly on unseen data.


##### Bias-Variance Tradeoff:
The bias-variance tradeoff is a fundamental concept in Machine Learning that involves finding the right balance between bias and variance to minimize the overall prediction error. Increasing model complexity reduces bias but increases variance, while decreasing complexity reduces variance but increases bias. The goal is to find a model complexity that achieves the lowest possible bias and variance, leading to optimal performance on unseen data.


 ## Regularization.

 The technique that can help prevent overfitting in Machine Learning models 

Regularization is a technique used to prevent overfitting by adding a penalty term to the model's loss function, which penalizes large coefficients or model complexity. This penalty encourages the model to learn simpler patterns and reduces its sensitivity to noise or irrelevant features in the training data.

Common types of regularization include L1 regularization (Lasso), which adds the absolute values of the coefficients to the loss function, and L2 regularization (Ridge), which adds the squared values of the coefficients. Both types of regularization help to constrain the model's complexity and reduce overfitting.

## Feature Scaling:

#### Concept:
Feature scaling is a technique used to standardize the range of independent variables or features in the data. It helps to bring all features to the same scale, preventing some features from dominating others during model training.
#### Example: 
Consider a dataset containing two features: age (ranging from 0 to 100) and income (ranging from 0 to 100,000). Feature scaling would ensure that both age and income are on the same scale, such as between 0 and 1, making it easier for the model to learn patterns from both features equally.
## Regularization:

#### Concept:
Regularization is a technique used to prevent overfitting by adding a penalty term to the model's loss function. This penalty discourages complex models with large coefficients, encouraging simpler models that generalize better to new data.
#### Example: 
In linear regression, regularization adds a penalty term to the loss function, such as the sum of squared coefficients (L2 regularization) or the sum of absolute coefficients (L1 regularization). This encourages the model to choose simpler coefficients, preventing overfitting.
## Data Augmentation:

#### Concept: 
Data augmentation is a technique used to artificially increase the size of the training dataset by applying transformations to the existing data samples. It helps to create variations of the original data, making the model more robust and reducing overfitting.
#### Example:
In image classification, data augmentation techniques may include rotating, flipping, or cropping images, changing their brightness or contrast, or adding noise. By generating new images with these transformations, the model learns to recognize objects from different perspectives, leading to better generalization.
## Ensemble Methods:

#### Concept:
Ensemble methods combine predictions from multiple individual models to make a final prediction. By leveraging the collective wisdom of diverse models, ensemble methods often result in more accurate and robust predictions compared to individual models.
#### Example:
Random Forest is an ensemble method that combines the predictions of multiple decision trees. Each decision tree is trained on a random subset of the data and makes its own prediction. The final prediction of the Random Forest is the average (for regression) or the majority vote (for classification) of all individual tree predictions.
